```python
# Inheritance in Python 2
class parent(object):
    ## 1. Implicit Inheritance
    def implicit(self):
        print("PARENT implicit()")

    ## 2. Override explicitly
    def override(self):
        print("PARENT override()")

    ## 3. Alter before or after (Using super)
    def altered(self):
        print("PARENT altered()")

    def __init__(self, last_name = None, *para, **mpara):
        if last_name is None:
            pass
        else:
            super(parent, self).__init__(*para, **mpara)
            self.last_name = last_name

class ntr(object):
    def __init__(self, blood_type = None, *para, **mpara):
        if blood_type is None:
            pass
        else:
            super(ntr, self).__init__(*para, **mpara)
            self.blood_type = blood_type 

class child(parent, ntr):
    def override(self):
        print("CHILD override()")
    def altered(self):
        print("BEFORE CHILD altered()")
        super(child,self).altered()
        print("AFTER CHILD altered()")
    ## 4. Multiple Inheritance
    def __init__(self, blood_type = None, last_name = None, age = None):
        if age and last_name and blood_type is None:
            pass
        else:
            super(child, self).__init__(blood_type = blood_type, last_name = last_name)
            self.age = age
    def show_up(self):
        print("age : " + str(self.age))
        print("last name : " + self.last_name)
        print("blood type : " + self.blood_type)

# Inheritance in Python 2
## 1. Implicit Inheritance
dad = parent()
son = child()

dad.implicit()
son.implicit()
print("*********************")

## 2. Override explicitly
dad.override()
son.override()
print("*********************")

## 3. Alter before or after (Using super)
dad.altered()
son.altered()
print("*********************")

## 4. Multiple Inheritance
whoreson = child("AB", "Greys", 18)
whoreson.show_up()
print("*********************")
```

```python
# Composition
class other(object):
    def override(self):
        print("OTHER override")

    def implicit(self):
        print("OTHER implicit")

    def altered(self):
        print("OTHER altered")

class another_child(object):
    def __init__(self):
        self.other = other()

    def override(self):
        print("CHILD override")

    def implicit(self):
        self.other.implicit()

    def altered(self):
        print("CHILD, before OTHER altered")
        self.other.altered()
        print("CHILD, after OTHER altered")
        
# Compostion
kid = another_child()
kid.override()
kid.implicit()
kid.altered()
```

```python
# Multiple Inheritance
class A(object):
    def __init__(self, x, *a, **b):
        super(A, self).__init__(*a, **b)
        self.x = x
        print "This is A" + str(self.x)

class Z(object):
    def __init__(self, *a, **b):
        print "This is Z"

# class B(Z):
class B(object):
    def __init__(self, v, *a, **b):
        super(B, self).__init__(*a, **b)
        self.v = v
        print "This is B" + str(self.v)

# class C(Z):
class C(object):
    def __init__(self, w, *a, **b):
        super(C, self).__init__(*a, **b)
        self.w = w
        print "This is C" + str(self.w)

# Class Q inherit from A -> B -> C
class Q(C,B,A):
    # the order is matter
    def __init__(self, x,v, w, var):
        # in any order 
        super(Q, self).__init__(x=x,v=v,w=w)
        self.var = var
        print "This is Q" + str(self.var)
a = A(1)
b = B(2)
c = C(3)
print("---")
q = Q(1, 2, 3, "asd")
```
